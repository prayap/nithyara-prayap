```python
#convert to strings
# brandstring=list((map(lambda x: str(x), brands_keywords_list)))
import re
import string
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import functools
import operator
import emoji
#NLP processing create functions to remove punctuation, extrawhite space, hex values, stopwords, emojis, links
def removePunct(sent):    
  sent = "".join([char for char in sent if char not in string.punctuation]) 
  return sent

def removeExtraWhitespaces(sent):
  sent = re.sub('\s+', ' ', sent).strip()
  return sent

def removeHex(sent):
  sent = re.sub(r"\\[a-z][a-z]?[0-9]+",r'', sent)
  return sent

def removestopwords(sent):
  stop_words=STOPWORDS
  word_list=sent.split()
  clean_sentence=' '.join([w for w in word_list if w.lower() not in stop_words])
  return(clean_sentence)

def removeemoji(sent):
  new_text = re.sub(emoji.get_emoji_regexp(), r"", sent)
  return new_text 
def removelinks(sent):
  string = re.sub(r'http//t.co\/[^\s]+', '', sent, flags=re.MULTILINE)
  return(string)
```

```python
#performing NLP on text to convert caption to simple text
combineddf['sim_text']=combineddf['Post Caption'].apply(lambda x :removeExtraWhitespaces(x))
combineddf['sim_text']=combineddf['sim_text'].apply(lambda x :removePunct(x) )
combineddf['sim_text']=combineddf['sim_text'].apply(lambda x :removeHex(x) )
combineddf['sim_text']=combineddf['sim_text'].apply(lambda x :removestopwords(x) )
combineddf['sim_text']=combineddf['sim_text'].apply(lambda x :removeemoji(x) )
combineddf['sim_text']=combineddf['sim_text'].apply(lambda x :removelinks(x) )
```
