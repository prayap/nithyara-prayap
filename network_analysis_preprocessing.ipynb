```python
#processing json files
joined_post_json_1 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'joined_post_json_1.csv')
joined_post_json_1 = joined_post_json_1[joined_post_json_1['Sponsorship Label'] == 1]
b = joined_post_json_1['File Contents'].str.findall(r'@[a-zA-Z0-9._]+')
account_id = joined_post_json_1['Account ID'].tolist()
flat_list = [list(set(item)) for item in b]
brand_influencer_df = pd.DataFrame(
    {'Account ID': account_id,
     'Brands': flat_list})
brand_influencer_df_edit = brand_influencer_df.explode('Brands')
brand_influencer_df_edit['Brands'] = brand_influencer_df_edit['Brands'].apply(lambda x: str(x)[1:])
merge = brand_influencer_df_edit.merge(brands[['Account ID', 'Category']], how='left', left_on='Brands', right_on='Account ID')
merge[['Category']] = merge[['Category']].fillna('Other')
merge[['Category']] = merge[['Category']].replace('', 'Other')
merge = merge.dropna()
merge = merge.drop(['Account ID_y'], axis=1)
merge = merge.rename({'Category': 'Brand Category', 'Brands': 'Brand', 'Account ID_x': 'Influencer'}, axis=1)
merge1 = merge.drop_duplicates()
```
