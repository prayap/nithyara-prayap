## Joining Counters AND Create Consolidated CSV

```python
influencers_sponsored_mentioning_brand_1 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'influencers_sponsored_mentioning_brand_1.csv')
influencers_sponsored_mentioning_brand_2 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'influencers_sponsored_mentioning_brand_2.csv')
influencers_sponsored_mentioning_brand_3 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'influencers_sponsored_mentioning_brand_3.csv')
influencers_sponsored_mentioning_brand_4 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'influencers_sponsored_mentioning_brand_4.csv')
```

```python
influencers_sponsored_mentioning_brand_combined = pd.concat([influencers_sponsored_mentioning_brand_1, influencers_sponsored_mentioning_brand_2, influencers_sponsored_mentioning_brand_3, influencers_sponsored_mentioning_brand_4]).groupby(['username']).sum().reset_index()
brands = pd.read_csv('drive/Shareddrives/capstone_prayap_nithyara/brands_data.csv')
brand_usernames =list(brands['Account ID'])
influencers_sponsored_mentioning_brand_combined['username'] = influencers_sponsored_mentioning_brand_combined['username'].apply(lambda x: x[1:])
influencers_sponsored_mentioning_brand_combined = influencers_sponsored_mentioning_brand_combined[influencers_sponsored_mentioning_brand_combined['username'].isin(brand_usernames)]
influencers_sponsored_mentioning_brand_combined = influencers_sponsored_mentioning_brand_combined.drop(['Unnamed: 0'], axis=1)

influencers_sponsored_mentioning_brand_combined.to_csv('drive/Shareddrives/capstone_prayap_nithyara/influencers_sponsored_mentioning_brand_combined.csv')
```

```python
influencers_sponsored_mentioning_brand_merge = pd.read_csv('drive/Shareddrives/capstone_prayap_nithyara/influencers_mentioning_brand_merge.csv')
influencers_sponsored_mentioning_brand_merge = influencers_sponsored_mentioning_brand_merge.merge(influencers_sponsored_mentioning_brand_combined, how='left', left_on='Account ID', right_on='username')
influencers_sponsored_mentioning_brand_merge = influencers_sponsored_mentioning_brand_merge.drop(['username'], axis=1)
influencers_sponsored_mentioning_brand_merge = influencers_sponsored_mentioning_brand_merge.rename(columns={'count': 'Influencers Sponsored Mentioning Brand'})
influencers_sponsored_mentioning_brand_merge = influencers_sponsored_mentioning_brand_merge.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.1.1', 'Unnamed: 0.1.1.1'], axis=1)
influencers_sponsored_mentioning_brand_merge
```
