## Join Counters and Create Consolidated CSV

```python
brand_mentions_by_influencer_1 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'brand_mentions_by_influencer_1.csv')
brand_mentions_by_influencer_2 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'brand_mentions_by_influencer_2.csv')
brand_mentions_by_influencer_3 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'brand_mentions_by_influencer_3.csv')
brand_mentions_by_influencer_4 = pd.read_csv("drive/Shareddrives/capstone_prayap_nithyara/" + 'brand_mentions_by_influencer_4.csv')
```

```python
brand_mentions_by_influencer_4
```

```python
brand_mentions_by_influencer_combined = pd.concat([brand_mentions_by_influencer_1, brand_mentions_by_influencer_2, brand_mentions_by_influencer_3, brand_mentions_by_influencer_4]).groupby(['Account ID']).sum().reset_index()
influencers = pd.read_csv('drive/Shareddrives/capstone_prayap_nithyara/influencers_data.csv')
influencers_usernames =list(influencers['Account ID'])
brand_mentions_by_influencer_combined = brand_mentions_by_influencer_combined[brand_mentions_by_influencer_combined['Account ID'].isin(influencers_usernames)]
brand_mentions_by_influencer_combined = brand_mentions_by_influencer_combined.drop(['Unnamed: 0'], axis=1)
brand_mentions_by_influencer_combined
brand_mentions_by_influencer_combined.to_csv('drive/Shareddrives/capstone_prayap_nithyara/brand_mentions_by_influencer_combined.csv')
```

```python
brand_mentions_by_influencer_merge = pd.read_csv('drive/Shareddrives/capstone_prayap_nithyara/brand_mentions_by_influencer_combined.csv')
brand_mentions_by_influencer_merge = influencers.merge(brand_mentions_by_influencer_merge, how='left', left_on='Account ID', right_on='Account ID')
brand_mentions_by_influencer_merge = brand_mentions_by_influencer_merge.rename(columns={'count': 'Influencers Mentioning Brand'})
brand_mentions_by_influencer_merge = brand_mentions_by_influencer_merge.drop(['Unnamed: 0_y', 'Unnamed: 0_x'], axis=1)
brand_mentions_by_influencer_merge

```

```python
sponsored_brand_mentions_by_influencer_merge = pd.read_csv('drive/Shareddrives/capstone_prayap_nithyara/brand_mentions_by_influencer_merge.csv')
sponsored_brand_mentions_by_influencer_merge = sponsored_brand_mentions_by_influencer_merge.merge(sponsored_brand_mentions_by_influencer_combined, how='left', left_on='Account ID', right_on='Account ID')
sponsored_brand_mentions_by_influencer_merge = sponsored_brand_mentions_by_influencer_merge.drop(['Unnamed: 0'], axis=1)
sponsored_brand_mentions_by_influencer_merge

```
